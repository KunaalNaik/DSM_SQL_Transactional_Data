```python
import pandas as pd
import random

# Define months and operators
months = ['Jan', 'Feb', 'Mar']
operators = ['Operator A', 'Operator B']

# Initialize an empty list to store the data
data = []

# Generate data for each month and operator
for month in months:
    for operator in operators:
        # Generate random data for each KPI
        cycle_time = random.randint(40, 50)
        throughput = random.randint(500, 650)
        utilization_rate = random.uniform(75, 90)
        mtbf = random.randint(170, 230)
        mttr = random.uniform(1.5, 2.5)
        planned_downtime = random.randint(10, 25)
        unplanned_downtime = random.randint(5, 12)
        scrap_rate = random.uniform(2, 3.5)
        rework_rate = random.uniform(1, 2.5)
        first_pass_yield = 100 - scrap_rate - rework_rate
        acknowledgment_time = random.randint(4, 8)
        resolution_time = random.randint(20, 35)
        escalation_time = random.randint(12, 20)
        
        # Append data to the list
        data.append([month, operator, cycle_time, throughput, utilization_rate,
                     mtbf, mttr, planned_downtime, unplanned_downtime,
                     scrap_rate, rework_rate, first_pass_yield,
                     acknowledgment_time, resolution_time, escalation_time])

# Create a DataFrame from the list of data
df = pd.DataFrame(data, columns=['Month', 'Operator', 'Cycle Time (mins)', 'Throughput (units)', 'Utilization Rate (%)',
                                 'MTBF (hours)', 'MTTR (hours)', 'Planned Downtime (hours)', 'Unplanned Downtime (hours)',
                                 'Scrap Rate (%)', 'Rework Rate (%)', 'First-pass Yield (%)',
                                 'Acknowledgment Time (mins)', 'Resolution Time (mins)', 'Escalation Time (mins)'])

# Display the DataFrame
df
```
